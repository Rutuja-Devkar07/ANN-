1. Write a Python program to plot a few activation functions that are being used in neural networks. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Define input range
x = np.linspace(-10, 10, 1000)

# Activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def linear(x):
    return x

# Plotting
plt.figure(figsize=(12, 8))

# Sigmoid
plt.subplot(2, 2, 1)
plt.plot(x, sigmoid(x), label='Sigmoid', color='blue')
plt.title('Sigmoid Function')
plt.grid(True)

# Tanh
plt.subplot(2, 2, 2)
plt.plot(x, tanh(x), label='Tanh', color='green')
plt.title('Tanh Function')
plt.grid(True)

# ReLU
plt.subplot(2, 2, 3)
plt.plot(x, relu(x), label='ReLU', color='red')
plt.title('ReLU Function')
plt.grid(True)

# Linear
plt.subplot(2, 2, 4)
plt.plot(x, linear(x), label='Linear', color='purple')
plt.title('Linear Function')
plt.grid(True)

plt.tight_layout()
plt.show()
```
